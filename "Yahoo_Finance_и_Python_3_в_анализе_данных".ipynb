Получение данных о ценах акций.

```python
import yfinance as yf

import matplotlib.pyplot as plt


# Получение данных для акций Apple

data = yf.download('AAPL', '2020-01-01', '2022-12-31')


# Визуализация скорректированной цены закрытия

data.plot()

plt.title('Скорректированная цена закрытия Apple')

plt.xlabel('Дата')

plt.ylabel('Цена')

plt.show()
```

**Задача 2:** Визуализация данных
Шаг 1: Создание списка тикеров.

```python

tickers_list = ['AAPL', 'GOOG', 'MSFT']
```

Шаг 2: Получение и визуализация данных для нескольких акций.

```python
import pandas as pd


# Создание DataFrame для хранения данных

data = pd.DataFrame(columns=tickers_list)


# Получение данных для каждого тикера

for ticker in tickers_list:

    data[ticker] = yf.download(ticker, '2020-01-01', '2022-12-31')['Close']


# Визуализация данных

data.plot(figsize=(10, 7))

plt.title('Скорректированная цена закрытия')

plt.xlabel('Дата')

plt.ylabel('Цена')

plt.legend()

plt.show()
```
